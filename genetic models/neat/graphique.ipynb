```python
#Graphique

import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('donnees_evolution.csv')

# Calculer la Fitness Moyenne par gnration
fitness_moyenne_par_generation = df.groupby('generation')['fitness_max'].mean().reset_index()
fitness_max_par_generation = df.groupby('generation')['fitness_max'].max().reset_index()

# Crer un graphique
plt.figure(figsize=(10, 6))

plt.plot(fitness_moyenne_par_generation['generation'], fitness_moyenne_par_generation['fitness_max'], label='Fitness Moyenne')
plt.plot(fitness_max_par_generation['generation'], fitness_max_par_generation['fitness_max'], label='Fitness max')

plt.xlabel('Gnration')
plt.ylabel('Fitness')
plt.legend()
plt.title('volution de la Fitness Max et de la Fitness Moyenne par Gnration')
plt.grid(True)
plt.show()
```
